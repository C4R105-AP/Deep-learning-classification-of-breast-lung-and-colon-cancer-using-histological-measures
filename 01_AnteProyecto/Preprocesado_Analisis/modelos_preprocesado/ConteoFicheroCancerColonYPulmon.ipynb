```python
import pandas as pd
import os
import mimetypes
import matplotlib.pyplot as plt

# Ajustar el tamaño de la gráfica
plt.figure(figsize=(10, 6))
```

```python
#Metodo para contar el número de archivos en una carpeta y sus subcarpetas
def contar_archivos_y_tipos(carpeta):
    total_archivos = 0
    tipos_archivos = {}

    for ruta_actual, carpetas, archivos in os.walk(carpeta):
        for archivo in archivos:
            total_archivos += 1
            ruta_completa = os.path.join(ruta_actual, archivo)
            tipo_archivo, _ = mimetypes.guess_type(ruta_completa)
            tipo_archivo = tipo_archivo or 'Tipo desconocido'

            if tipo_archivo in tipos_archivos:
                tipos_archivos[tipo_archivo] += 1
            else:
                tipos_archivos[tipo_archivo] = 1

    return total_archivos, tipos_archivos
```

```python
#Metodo para contar el número de archivos por carpeta
def contar_archivos_por_carpeta(carpeta):
    archivos_por_carpeta = {}

    archivos_por_carpeta = {}

    for ruta_actual, carpetas, archivos in os.walk(carpeta):
        nombre_carpeta = os.path.basename(ruta_actual)
        archivos_por_carpeta[nombre_carpeta] = len(archivos)

    return archivos_por_carpeta
```

```python
carpeta_inicio = r'C:\Users\lacer\OneDrive\Escritorio\ejercicios\TFM2'  #Escribir aquí ubicación de la carpeta de imagenes
total, tipos = contar_archivos_y_tipos(carpeta_inicio)

print("Total de archivos:", total)
print("Tipos de archivos:")

for tipo, cantidad in tipos.items():
    print(f"- {tipo}: {cantidad}")
```

```python
def contar_archivos_por_subcarpeta(ruta):
    # Inicializar un diccionario para almacenar el recuento de archivos por subcarpeta
    recuento_archivos = {}

    # Recorrer las subcarpetas y archivos en la ruta especificada
    for directorio_raiz, directorios, archivos in os.walk(ruta):
        # Obtener el nombre de la subcarpeta actual
        nombre_subcarpeta = os.path.basename(directorio_raiz)

        # Obtener el recuento de archivos en el directorio raíz (incluyendo las subcarpetas)
        recuento = sum([len(ficheros) for _, _, ficheros in os.walk(directorio_raiz)])

        # Agregar el recuento al diccionario, utilizando el nombre de la subcarpeta como clave
        recuento_archivos[nombre_subcarpeta] = recuento

    # Devolver el diccionario con el recuento de archivos por subcarpeta
    return recuento_archivos
```

```python
recuento = contar_archivos_por_subcarpeta(carpeta_inicio)
# Imprimir el recuento de archivos por subcarpeta
for subcarpeta, num_archivos in recuento.items():
    print(f"Carpeta {subcarpeta} - Número de archivos: {num_archivos}")
```

```python
# Etiquetas de las barras
labels = ['Benignas', 'Cancerígenas']

# Alturas de las barras
heights = [recuento["colon_n"], recuento["colon_aca"]]

# Colores de las barras
colors = ['#2ecc71', '#c0392b']  # Verde para benigno, rojo para cáncer

# Crear el gráfico de barras
plt.bar(labels, heights, color=colors)

# Configurar el estilo del título en negrita
title_font = {'fontweight': 'bold'}

# Título y etiquetas de los ejes
plt.title('Células cancerígenas vs benignas de colon', **title_font)
plt.xlabel('Subcarpeta')
plt.ylabel('Total de archivos')

# Mostrar el gráfico
plt.show()
```
