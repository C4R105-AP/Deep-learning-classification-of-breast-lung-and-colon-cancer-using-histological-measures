```python
import pandas as pd
import os
import mimetypes
import random
import cv2
import numpy as np
from google.colab import drive
```

```python
# Ruta de la carpeta principal que contiene las subcarpetas "0" y "1"
drive.mount('/content/drive')

ruta_principal = '/content'
```

```python
num_imagenes_seleccionadas = 5000
paths_imagenes_0 = []
paths_imagenes_1 = []
num_images_0=0
num_images_1=0

def obtener_imagenes(carpeta, numImages):
    paths_imagenes = []
    for archivo in os.listdir(carpeta):
        if (numImages<num_imagenes_seleccionadas):
            path_imagen = os.path.join(carpeta, archivo)
            if os.path.isfile(path_imagen):
                #print(paths_imagenes)
                paths_imagenes.append(path_imagen)
                numImages+=1
        else:
            break
    return paths_imagenes, numImages

# Recorrer la carpeta principal
for carpeta_principal in os.listdir(ruta_principal):
    ruta_carpeta_principal = os.path.join(ruta_principal, carpeta_principal)
    if os.path.isdir(ruta_carpeta_principal):
        carpeta_0 = os.path.join(ruta_carpeta_principal, "0")
        carpeta_1 = os.path.join(ruta_carpeta_principal, "1")
        if os.path.isdir(carpeta_0):
            path_images, num_images_0=obtener_imagenes(carpeta_0, num_images_0)
            paths_imagenes_0.extend(path_images)
        if os.path.isdir(carpeta_1):
            path_images, num_images_1=obtener_imagenes(carpeta_1, num_images_1)
            paths_imagenes_1.extend(path_images)



```

```python
# Mostrar la cantidad de imágenes seleccionadas
print(f"Imágenes de la carpeta '0' seleccionadas: {len(paths_imagenes_0)}")
print(f"Imágenes de la carpeta '1' seleccionadas: {len(paths_imagenes_1)}")
print(paths_imagenes_1[0])
```

```python
# Lista para almacenar los datos de las imágenes
X_train = []

# Iterar sobre los paths de las imágenes y cargar los datos
for path_imagen in paths_imagenes_0:
    with open(path_imagen, 'rb') as f:
        imagen_data = f.read()
    X_train.append(imagen_data)
```
