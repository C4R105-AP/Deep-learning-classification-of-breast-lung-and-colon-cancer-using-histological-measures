```python
import cv2
import os
import pandas as pd
import numpy as np

# Define the image dimensions
IMG_HEIGHT = 768
IMG_WIDTH = 768
N_CHANNELS = 3  # For color images

# Define the base directories
base_dir = 'C:\Escritorio_10_9_2022\Master Artificial Intelligence\Modulo 7\data\lung_colon_image_set'
lung_dir = os.path.join(base_dir, 'C:\Escritorio_10_9_2022\Master Artificial Intelligence\Modulo 7\data\lung_colon_image_set\lung_image_sets')
colon_dir = os.path.join(base_dir, 'C:\Escritorio_10_9_2022\Master Artificial Intelligence\Modulo 7\data\lung_colon_image_set\colon_image_sets')

# Define the categories for both lung and colon
lung_categories = ['lung_aca', 'lung_n', 'lung_scc']
colon_categories = ['colon_aca', 'colon_n']

def create_dataframe(image_dir, categories):
    data = []

    # iterate over the categories
    for category in categories:
        path = os.path.join(image_dir, category)
        class_num = categories.index(category)  # assign an integer label to this category

        # iterate over all images in the sub-directory
        for img in os.listdir(path):
            try:
                # read the image and resize
                img_array = cv2.imread(os.path.join(path, img))
                # OpenCV reads images in BGR format, convert to RGB
                img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)

                # append the image data and class label to the dataset
                data.append([img_array, class_num])
            except Exception as e:
                print(e)

    # Create a DataFrame
    df = pd.DataFrame(data, columns=['image', 'label'])
    return df

# Creating the dataframes
lung_df = create_dataframe(lung_dir, lung_categories)
colon_df = create_dataframe(colon_dir, colon_categories)

lung_df['area'] = 'lung'
colon_df['area'] = 'colon'

# Concatenating the lung and colon dataframes into one
full_df = pd.concat([lung_df, colon_df])
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from torchvision import transforms

# Normalize the images (0-255 -> 0-1)
#full_df['image'] = full_df['image'].apply(lambda x: x / 255)

# Split data into training and testing sets
train_df, test_df = train_test_split(full_df, test_size=0.2, random_state=42)

class ImageDataset(Dataset):
    def __init__(self, df):
        self.df = df

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        return torch.tensor(row['image']).permute(2, 0, 1), torch.tensor(row['label'])

# Convert DataFrames into Dataset
train_dataset = ImageDataset(train_df)
test_dataset = ImageDataset(test_df)

# Create DataLoaders
batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
import torch.nn as nn
import torch.nn.functional as F

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(N_CHANNELS, 32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.fc1 = nn.Linear(64*192*192, 128)
        self.fc2 = nn.Linear(128, len(lung_categories + colon_categories))

    def forward(self, x):
        x = F.max_pool2d(F.relu(self.conv1(x)), 2)
        x = F.max_pool2d(F.relu(self.conv2(x)), 2)
        x = x.view(x.size(0), -1)  # Flatten layer
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)

# Instantiate the network
net = Net()
```

```python
import torch.optim as optim

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
```

```python
# Number of epochs to train for
num_epochs = 10

# Move tensors to the configured device
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
net = net.to(device)

# Begin training
for epoch in range(num_epochs):
    running_loss = 0.0
    for i, data in enumerate(train_loader, 0):
        inputs, labels = data[0].to(device), data[1].to(device)

        # Zero the parameter gradients
        optimizer.zero_grad()

        # Forward + backward + optimize
        outputs = net(inputs.float())
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        # print statistics
        running_loss += loss.item()
        if i % 2000 == 1999:  # print every 2000 mini-batches
            print('[%d, %5d] loss: %.3f' %
                  (epoch + 1, i + 1, running_loss / 2000))
            running_loss = 0.0

print('Finished Training')
```
